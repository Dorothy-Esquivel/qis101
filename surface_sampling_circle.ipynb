```python
"""surface_sampling_circle.py"""
# Cell 1

from __future__ import annotations

import typing

import matplotlib.pyplot as plt
import numpy as np
import numpy.random as random
from matplotlib.markers import MarkerStyle

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

%matplotlib widget

def plot_incorrect(ax: Axes) -> None:
    """
    Plots the incorrect sampling of points on a circle

    Args:
        ax: Matplotlib axes object on which to plot
    """
    num_samples = 10000

    # Generating torodial angle and radius values
    v: NDArray[np.float_] = np.linspace(0, 2 * np.pi, num_samples)  # toroidal angle
    r: NDArray[np.float_] = random.rand(num_samples)

    # Calculating x and y coordinates using incorrect formuls
    x: NDArray[np.float_] = r * np.cos(v)
    y: NDArray[np.float_] = r * np.sin(v)

    pixel_size: float = (72 / ax.figure.dpi) ** 2  # type: ignore

    # Plotting the sampled points
    ax.scatter(x, y, marker=MarkerStyle("."), s=pixel_size)

    # Setting plot title, x-axis label, y-axis label, and aspect ratio
    ax.set_title("Incorrect Sampling")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    ax.set_aspect("equal")


def plot_correct(ax: Axes) -> None:
    """
    Plots the correct sampling of points on a circle

    Args:
        ax: Matplotlib axes object on which to plot
    """
    num_samples = 10000

    # Generating torodial angle and radius values
    v: NDArray[np.float_] = np.linspace(0, 2 * np.pi, num_samples)  # toroidal angle
    r: NDArray[np.float_] = random.rand(num_samples)

    # TODO: Fix these next two lines
    # Used Dave's code
    # Calculating x and y coordinates using correct formula
    x: NDArray[np.float_] = np.array(np.sqrt(r) * np.cos(v), dtype=np.float_)
    y: NDArray[np.float_] = np.array(np.sqrt(r) * np.sin(v), dtype=np.float_)

    pixel_size: float = (72 / ax.figure.dpi) ** 2  # type: ignore

    # Plotting the sampled points
    ax.scatter(x, y, marker=MarkerStyle("."), s=pixel_size)

    # Setting plot title, x-axis label, y-axis, and aspect ratio
    ax.set_title("Correct Sampling")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    ax.set_aspect("equal")


def main() -> None:
    """ Main function that generates and displays the plots of incorrect and correct sampling on a circle"""
    plt.figure(" ", figsize=(10, 6))
    
    # Plotting the incorrect and correct sampling on subplots
    plot_incorrect(plt.subplot(121))
    plot_correct(plt.subplot(122))
    
    plt.show()


main()
```
